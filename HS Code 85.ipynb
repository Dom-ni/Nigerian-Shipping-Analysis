```python
# Loading the second dataset 
HS_code1 = pd.read_excel("HS CODE 85 JAN -DEC 2017_1.xlsx")
```

```python
HS_code1.info()
```

```python
HS_code1.isnull().sum()
```

```python
# Creating a function to automatically perform data cleaning
def clean_data(df, change_datatype=True, fill_missing=True):
    if change_datatype:
        df['Receipt Date'] = pd.to_datetime(df['Receipt Date'], dayfirst=True)
        df['Reg. Date'] = pd.to_datetime(df['Reg. Date'], dayfirst=True)
    if fill_missing:
        df['Receipt Date'] = df['Receipt Date'].fillna(method='ffill')
        df['Container Size'].fillna('Not specified', inplace=True)
        df['Container Nbr'].fillna('Non specified', inplace=True)
        df['Receipt Number'].fillna("Not specified", inplace=True)
        df['Country  of Supply'].fillna('Not specified', inplace=True)
    return df
```

```python
# Applying the function to the dataset
Cleaned_HS_Code1 = clean_data(HS_code1, change_datatype=True, fill_missing=True)
print(Cleaned_HS_Code1)
```

```python
Cleaned_HS_Code1.isnull().sum()
```

```python
# Saving second cleaned dataset
Cleaned_HS_Code1.to_excel('Cleaned_HS Code 85 JAN-DEC 2017_1.xlsx', index=False)
```
